```python
img = cv2.imread('/content/gdrive/My Drive/opencv colab/img/milkdrop.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR 이미지를 RGB 이미지로 변환

gImg= cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
plt.imshow(gImg,cmap='gray')
```

```python
ret, thresh = cv2.threshold(gImg, 127, 255, 0)

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

image = cv2.drawContours(img, contours, -1, (0, 0, 255), 1)
print("Segmented image")
cv2_imshow(thresh)
print("Contiur image")
cv2_imshow(image)
```

```python
print("Segmented image")
out = thresh.copy()
out = 255 - out
cv2_imshow(out)

print("Contiur image")
out2 = image.copy()
out2 = 255 - out2
cv2_imshow(out2)
```
