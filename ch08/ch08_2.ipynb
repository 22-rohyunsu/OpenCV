```python
from google.colab import drive
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
```

```python
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/opencv colab/img/scanning.jpg')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR 이미지를 RGB 이미지로 변환

gImg= cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gImg.shape)
plt.imshow(gImg,cmap='gray')
```

```python
gBin = np.zeros((gImg.shape[0], gImg.shape[1]), dtype = np.ubyte)

thershold = 150

for h in range(gImg.shape[0]):
  for w in range(gImg.shape[1]):
    if gImg[h,w] > thershold:
      gBin[h,w] = 255

cv2_imshow(gBin)
```

```python
kernel = np.ones((3, 3), np.uint8)
dilateImg = cv2.dilate(gBin, kernel, iterations=1)
cv2_imshow(dilateImg)
```

```python
kernel = np.ones((3,3), np.uint8)
erodeImg = cv2.erode(dilateImg, kernel, iterations=1)
cv2_imshow(erodeImg)
```
