```python
noiseImg = gImg.copy()
```

```python
from random import randint
```

```python
# 10% noise
noiseRatio = 10
whiteMax = noiseRatio/2
blackMax = noiseRatio
```

```python
noiseImg = gImg.copy()

for h in range(0, gImg.shape[0]):
  for w in range(0, gImg.shape[1]):
    val = randint(0, 100)
    if (val < whiteMax):
      noiseImg[h, w] = 255
    elif (val < blackMax):
      noiseImg[h, w] = 0

cv2_imshow(noiseImg)
```

```python
outImg = cv2.medianBlur(noiseImg, 3)
cv2_imshow(outImg)
```

```python
outImg2 = cv2.medianBlur(noiseImg, 5)
cv2_imshow(outImg2)
```
